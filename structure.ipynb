```python
def normalize_car_data(df, detail_col, overview_col, feature_col, specs_col):
   
    def convert_to_dict(cell):
        if isinstance(cell, str):
            try:
                return ast.literal_eval(cell)
            except (ValueError, SyntaxError):
                return cell
        return cell
    
    for col in [detail_col, overview_col, feature_col, specs_col]:
        df[col] = df[col].apply(convert_to_dict)
    
    car_details_df = pd.json_normalize(df[detail_col])
    car_overview_df = pd.json_normalize(df[overview_col])
    car_features_df = pd.json_normalize(df[feature_col])
    car_specs_df = pd.json_normalize(df[specs_col])
    def extract_key_value_dicts(details):
        if isinstance(details, list):
            return {item['key']: item['value'] for item in details}
        else:
            return {}

    car_top_df = car_overview_df['top'].apply(extract_key_value_dicts).apply(pd.Series)
    Car_df = pd.concat([car_details_df, car_top_df], axis=1)
    
    def count_features(features):
        if isinstance(features, list):
            return len(features)
        else:
            return 0
    car_features_df['top_features_count'] = car_features_df['top'].apply(count_features)
    car_features_df_data = pd.json_normalize(car_features_df['data'])

    categories = ['comfort', 'interior', 'exterior', 'safety', 'entertainment']
    for i, category in enumerate(categories):
        car_features_df_data[f'{category}_count'] = pd.json_normalize(car_features_df_data[i])['list']
        car_features_df_data[f'{category}_count'] = car_features_df_data[f'{category}_count'].apply(count_features)

    car_features_df_data.drop(columns=[0, 1, 2, 3, 4], inplace=True)
    features_df = pd.concat([car_features_df['top_features_count'], car_features_df_data], axis=1)
    Car_df = pd.concat([Car_df, car_features_df], axis=1)
    
    car_specs_top_df = car_specs_df['top'].apply(extract_key_value_dicts).apply(pd.Series)
    car_specs_data_df = pd.json_normalize(car_specs_df['data'])

    car_specs_data_df_engine = pd.json_normalize(car_specs_data_df[0])['list'].apply(extract_key_value_dicts).apply(pd.Series)
    car_specs_data_df_other = pd.json_normalize(car_specs_data_df[1])['list'].apply(extract_key_value_dicts).apply(pd.Series)
    car_specs_data_df_miss = pd.json_normalize(car_specs_data_df[2])['list'].apply(extract_key_value_dicts).apply(pd.Series)

    specs_df = pd.concat([car_specs_top_df, car_specs_data_df_engine, car_specs_data_df_other, car_specs_data_df_miss], axis=1)
    Car_df = pd.concat([Car_df, specs_df], axis=1)
    
    return Car_df
```

```python
che = pd.read_excel('chennai_cars.xlsx')
che_df = normalize_car_data(che, 'new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs')
che_df['City'] = 'Chennai'
che_df.head()
```

```python
kol = pd.read_excel('kolkata_cars.xlsx')
kol_df = normalize_car_data(kol, 'new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs')
kol_df['City'] = 'Kolkata'
kol_df.head()
```

```python
all_cities_car_data = [blr_df, che_df, dl_df, hyd_df, jai_df, kol_df]

def rename_duplicate_columns(df):
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        count = 0
        for i in range(len(cols)):
            if cols[i] == dup:
                count += 1
                if count > 1:
                    cols[i] = f"{dup}_{count-1}"
    df.columns = cols
    return df
    
all_cities_car_data = [rename_duplicate_columns(df) for df in all_cities_car_data]
all_columns = set()
for df in all_cities_car_data:
    all_columns.update(df.columns)

for df in all_cities_car_data:
    missing_cols = all_columns - set(df.columns)
    for col in missing_cols:
        df[col] = pd.NA 
all_cities_car_data = [df.reset_index(drop=True) for df in all_cities_car_data]
combined_df = pd.concat(all_cities_car_data, axis=0, join='outer', ignore_index=True)

print(combined_df.shape)
combined_df.head() 
```

```python
combined_df.to_excel('concat_cities_car_data.xlsx', index=False)
```
